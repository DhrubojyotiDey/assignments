```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f'The {self.name_of_vehicle} has a seating capacity of {capacity}.'
    
my_car = Car('Sedan', 120, 60)
print(my_car.seating_capacity(5))  # prints: The Sedan has a seating capacity of 5.

'''In this example, ‘Sedan’ is the name_of_vehicle, 120 is the max_speed, 60 is the average_of_vehicle, and 5 is the capacity.
You can replace these with the values for your specific car. 
The seating_capacity method will return a string that includes the name of the vehicle and its seating capacity.'''
```

```python
'''In Object-Oriented Programming, **multiple inheritance** is a feature where a class can inherit properties and methods
from more than one parent class. This allows for the creation of more complex classes through the combination of simpler ones.

Here's an example of multiple inheritance in Python:'''


class Parent1:
    def method1(self):
        print("Method from Parent1")

class Parent2:
    def method2(self):
        print("Method from Parent2")

class Child(Parent1, Parent2):
    def child_method(self):
        print("Method from Child")

# Create an instance of Child
child = Child()

# Call methods
child.method1()  # prints: Method from Parent1
child.method2()  # prints: Method from Parent2
child.child_method()  # prints: Method from Child

'''
In this example, `Child` is a class that inherits from both `Parent1` and `Parent2`, so an object of type 
`Child` can use methods from both parent classes as well as its own methods. This is multiple inheritance.'''
```

```python
'''A **getter** method is used to retrieve the value of an attribute. In Python, we use the `@property` decorator
to define a method as a getter.
A **setter** method is used to set or update the value of an attribute. In Python, we use the `@<attribute>.setter`
decorator to define a method as a setter.

Here's an example of a class with getter and setter methods:

'''
class Employee:
    def __init__(self, name=None):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise ValueError("Name must be a string.")
        self._name = value
'''

In this example, `name` is an attribute of the `Employee` class. The `name` method is a getter that returns the
value of `_name`, and the `name` setter method sets the value of `_name`. The setter also includes a check to make
sure that `value` is a string, and raises a `ValueError` if it's not.
You can use this class like this:

'''
emp = Employee()
emp.name = "Alice"  # sets the name
print(emp.name)  # gets the name, prints: Alice

'''
This way, you have control over how an attribute is set and returned, which can help prevent bugs and
make your code easier to understand.'''
```

```python
'''In Object-Oriented Programming, **method overriding** is a feature that allows a subclass to provide
a different implementation of a method that is already defined in its superclass. It is used when a child
class has the same method name as the parent class, and you want the child class's method to be executed 
instead of the parent's method.

Here's an example of method overriding in Python:
'''

class Animal:
    def sound(self):
        return "This is how an animal sounds"

class Dog(Animal):
    def sound(self):
        return "Dogs bark"

animal = Animal()
print(animal.sound())  # prints: This is how an animal sounds

dog = Dog()
print(dog.sound())  # prints: Dogs bark
'''

In this example, `Dog` is a subclass of `Animal`, and both classes have a method named `sound`.
When we call `sound` on an instance of `Dog`, Python uses the `sound` method defined in the `Dog` 
class, not the one in `Animal`. This is method overriding.'''
```
